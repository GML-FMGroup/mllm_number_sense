```python
from PIL import Image, ImageDraw, ImageFont
import json
import os
import shutil

index = 0
file_names = ["Synthetic Mathematical Dataset", "Real-World Dataset"]
json_names = ["/Synthetic.json", "/Real.json"]
file_name = file_names[index]
json_name= json_names[index]

# 设置 JSON 文件路径
json_path = r"../Only_data/" + file_name + json_name
root_dir = r"../"

# 载入 JSON 文件
with open(json_path, 'r', encoding='utf-8') as f:
    data = json.load(f)

# 设置新的目录路径
output_dir = file_name

# 如果目录存在，先清空目录
if os.path.exists(output_dir):
    shutil.rmtree(output_dir)  # 删除目录及其内容

# 创建新的空目录
os.makedirs(output_dir, exist_ok=True)

# 使用较大的字体（可以替换为自己本地的字体文件路径）
font_path = "arial.ttf"  # 这里你可以替换为自己系统中的字体文件路径
# font = ImageFont.truetype(font_path, 17)  # 设置字体大小为36

# 处理每个数据条目
for entry in data["data"]:
    # 读取原始图片
    original_image = Image.open(os.path.join(root_dir, entry["image_path"]))
    base_name = entry["image_path"].split("/")[-1]
    cur_cls = entry["Major_categories"]
    # if cur_cls != "Scale":
    #     continue
    # 创建白色背景图片
    white_image = Image.new("RGB", original_image.size, (255, 255, 255))

    # 拼接图像（原图在左，白色背景在右）
    new_image = Image.new("RGB", (original_image.width + white_image.width, original_image.height))
    new_image.paste(original_image, (0, 0))
    new_image.paste(white_image, (original_image.width, 0))

    # 在白色背景上添加答案文本
    draw = ImageDraw.Draw(new_image)
    
    entry_text = entry["text"]  # 获取文本
    words = entry_text.split(" ")  # 按空格分割文本

    # 每遇到第三个空格，将空格替换为换行符
    for i in range(2, len(words), 4):  # 从索引2开始，每隔3个词处理一次
        words[i] += "\n"  # 在单词后添加换行符

    # 将修改后的词列表重新拼接成字符串
    modified_text = " ".join(words) + "\n"
    
    op = entry["option"].replace("(c)", "\n(c)")
    text = modified_text+ op +"\n\n\n" + entry["answer"]
    
    # 计算文本的边界框以使其居中
    # bbox = draw.textbbox((0, 0), text, font=font)
    bbox = draw.textbbox((0, 0), text)

    
    text_width, text_height = bbox[2] - bbox[0], bbox[3] - bbox[1]
    
    # 计算文本居中位置
    # text_position = (original_image.width + (white_image.width - text_width) // 2, (original_image.height - text_height) // 2)
    text_position = (original_image.width,0)
    
    
    # 绘制文本，颜色为红色
    # draw.text(text_position, text, font=font, fill=(255, 0, 0))  # 红色文字
    draw.text(text_position, text, fill=(255, 0, 0))  # 红色文字

    cur_dir = os.path.join(output_dir, cur_cls)
    if os.path.exists (cur_dir) == False:
        os.mkdir(cur_dir)
    # 保存处理后的图片到新目录
    output_path = os.path.join(cur_dir, base_name)
    new_image.save(output_path)

print("处理完成！")
```
